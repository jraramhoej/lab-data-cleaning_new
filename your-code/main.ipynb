#### 5. Rename Id column to postId and OwnerUserId to userId

```python
# rename column using rename()
posts = posts.rename(columns={"Id":"postId", "OwnerUserId": "userId"})

# check df
posts["userId"].head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
# choose columns for new dfs
new_users = users[["userId", "Reputation", "Views", "UpVotes", "DownVotes"]]
new_posts = posts[["postId", "Score", "userId", "ViewCount", "CommentCount"]]

# check dfs
display(new_users.head())
display(new_posts.head())
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
# check for columns with the same name
print(np.intersect1d(users.columns, posts.columns))

# make an inner merge
merged_dfs = posts.merge(users, how="inner", left_on="userId", right_on="userId")

# check merged dfs
merged_dfs.head()
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain.

```python
# For the columns with strings, we can change NaN to "NA" in order to enable string operations on those columns

merged_dfs["Body"].fillna('NA', inplace=True)
merged_dfs["Title"].fillna('NA', inplace=True)
merged_dfs["Tags"].fillna('NA', inplace=True)
merged_dfs["OwnerDisplayName"].fillna('NA', inplace=True)
merged_dfs["LastEditorDisplayName"].fillna('NA', inplace=True)
merged_dfs["DisplayName"].fillna('NA', inplace=True)
merged_dfs["Location"].fillna('NA', inplace=True)
merged_dfs["AboutMe"].fillna('NA', inplace=True)
merged_dfs["WebsiteUrl"].fillna('NA', inplace=True)
merged_dfs["ProfileImageUrl"].fillna('NA', inplace=True)

```
