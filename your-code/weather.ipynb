## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
```

```python
# import csv
weather = pd.read_csv("/Users/rickardramhoj/ironhack_assignments/lab-data-cleaning/weather-raw.csv")

# check df
weather.head()
```

```python
# melt the days of the month to one column

# columns to keep
columns_to_keep = ["id", "year", "month", "element"]

weather = pd.melt(weather, id_vars = columns_to_keep)

# change names of new columns 
weather = weather.rename(columns={"variable": "day", "value": "temperature"})

# check df
weather.head()
```

```python
# put together year month and day to one datetime column
weather["date"] = weather["day"].str.replace('d', '') + "/" + weather["month"].astype(str) + "/" + weather["year"].astype(str)

# change the datatype of the date column
weather["date"] = pd.to_datetime(weather["date"], format='%d/%m/%Y', errors='coerce')

# drop year, month, day columns
weather.drop(labels=["year", "month", "day"], axis=1, inplace=True)

#weather.dtypes
weather.head(50)
```
